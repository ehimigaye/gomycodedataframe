```python
#Question 1:
#Creons d' abord la dataframe :
import pandas as pd #importons pandas 
import numpy as np#importants numpys pour les donnees nan
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],

'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],

'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],

'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
exam =  pd.DataFrame(exam_data)#creons la dataframe avec uniquement le dictionnaire exam_data
exam#verifions
labels = pd.DataFrame(labels, columns = ["labels"])#creons la series labels 
exam = pd.concat([labels, exam], axis = 1)#Concatenons le dataframe et la serie pour avoir la dataframe demand√©e
exam10 = exam.set_index('labels')#transformons la colonne label en index
exam10
exam10.head()# selectionnons les 5 premieres lignes du dataframe

```
